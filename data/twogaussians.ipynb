```python
import numpy as np
from ligeor import TwoGaussianModel
import matplotlib.pyplot as plt
```

```python
C = 1.
mu1 = 0.
sigma1 = 0.015
d1 = 0.5
mu2 = 0.42
sigma2 = 0.01
d2 = 0.35
Aell = 0.05
sigma_noise = 0.01
```

```python
phases = np.random.uniform(low=-0.5,high=0.5,size=1000)
phases = phases[np.argsort(phases)]
```

```python
np.random.seed(42)
noise = np.random.normal(loc=0., scale=sigma_noise, size=len(phases))
```

```python
const_model = TwoGaussianModel.const(phases, C) + noise
ce_model = TwoGaussianModel.ce(phases, C, Aell, mu1) + noise
cg_model = TwoGaussianModel.cg(phases, C, mu1, d1, sigma1) + noise
cge_model = TwoGaussianModel.cge(phases, C, mu1, d1, sigma1, Aell) + noise
cg12_model = TwoGaussianModel.cg12(phases, C, mu1, d1, sigma1, mu2, d2, sigma2) + noise
cg12e1_model = TwoGaussianModel.cg12e1(phases, C, mu1, d1, sigma1, mu2, d2, sigma2, Aell) + noise
cg12e2_model = TwoGaussianModel.cg12e2(phases, C, mu1, d1, sigma1, mu2, d2, sigma2, Aell) + noise
```

```python
for model, label in zip([const_model, ce_model, cg_model, cge_model, cg12_model, cg12e1_model, cg12e2_model],
                 ['C', 'CE', 'CG', 'CGE', 'CG12', 'CG12E1', 'CG12E2']):
    plt.plot(phases, model, '.', label=label)
    plt.legend()
    plt.show()
```

```python
np.savetxt('const.csv', np.array([phases, const_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('cg.csv', np.array([phases, cg_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('ce.csv', np.array([phases, ce_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('cge.csv', np.array([phases, cge_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('cg12.csv', np.array([phases, cg12_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('cg12e1.csv', np.array([phases, cg12e1_model, np.abs(noise)]).T, delimiter=',')
np.savetxt('cg12e2.csv', np.array([phases, cg12e2_model, np.abs(noise)]).T, delimiter=',')
```
